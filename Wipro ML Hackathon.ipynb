# Check for shape and columns of dataset

```python
df.shape
```

```python
df.columns
```

```python
X_train = df.drop(['Clearsky DHI','Clearsky DNI','Clearsky GHI'],axis=1)
```

```python
y1=df['Clearsky DHI']
y2=df['Clearsky DNI']
y3=df['Clearsky GHI']
```

```python
#model training for train dataset

from sklearn.ensemble import RandomForestRegressor
```

```python
model1= RandomForestRegressor()
model2=RandomForestRegressor()
model3=RandomForestRegressor()
```

```python
model1.fit(X_train,y1)
model2.fit(X_train,y2)
model3.fit(X_train,y3)
```

## Repeat with the test data

```python
df_test.head()
```

```python
X_test=df_test.drop(['Clearsky DHI','Clearsky DNI','Clearsky GHI'],axis=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
pred1=model1.predict(X_test).astype('int32')
pred2=model2.predict(X_test).astype('int32')
pred3=model3.predict(X_test).astype('int32')
```

```python
df_test['Clearsky DHI']=pred1
df_test['Clearsky DNI']=pred2
df_test['Clearsky GHI']=pred3
```

```python
submission = df_test[['Clearsky DHI','Clearsky DNI','Clearsky GHI']]
```

```python
submission.to_csv('wipro_submission.csv',index=False)
```
